```python
import pandas as pd

pd_reader = pd.read_csv("data.csv",encoding="ANSI")
display(pd_reader)
df = pd_reader
```

```python
df[df['業種']=='銀行業']
```

```python
rtn = df.loc[:, ['企業名','配当利回り']]
#display(rtn)
rtn = rtn.loc[:,'配当利回り']

print('平均配当利回り:', rtn.mean())
```

```python
sort = df.sort_values('利益成長',ascending=False)
#print(sort)
sort = sort.iloc[0:10]
top10 = sort.loc[:,'利益成長']

print('利益成長の上位10社の平均値:', top10.mean())
```

```python
amount = df.loc[:, ['企業名','売上高']]
amount = amount.loc[:,'売上高']

print('平均値：',amount.mean(), 
      '\n中央値：',amount.median())
```

```python
ROE = df.loc[:, 'ROE']
#print(ROE)
ROE_high = df[df.ROE > ROE.quantile(q=0.9)]
ROE_low  = df[df.ROE < ROE.quantile(q=0.1)]

print('ROE上位10％と下位10％の差：',ROE_high.loc[:, 'ROE'].mean() - ROE_low.loc[:, 'ROE'].mean())
```

```python
amount0 = df.loc[:, '売上高']

# 売上高上位10％の企業
high = df[df.売上高 > amount0.quantile(q=0.9)]
high = high.sort_values('売上高',ascending=False)
display(high)
```

```python
# 売上高下位10％の企業
low = df[df.売上高 < amount0.quantile(q=0.1)]
low = low.sort_values('売上高',ascending=False)
display(low)
```

```python
# T検定を行う
import math

high_mean = high.loc[:,'売上高'].mean()
high_var = high.loc[:,'売上高'].var()
n_x = 210

low_mean = low.loc[:,'売上高'].mean()
low_var = low.loc[:,'売上高'].var()
n_y = 210

T = (high_mean - low_mean) / math.sqrt(high_var/n_x + low_var/n_y)
print(T)

def student(t):
    if t > 2:
        print('有意差あり')
    else:
        print('有意差なし')
    return

student(T)
```
